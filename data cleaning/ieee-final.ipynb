```python
df1.drop(['Leader Year','Leader College'],axis=1,inplace=True)
df1.head()
```

```python
df1['Candidate Type']= 'team leader'
df1.head()
```

```python
df1 = df1.rename(columns={'Leader Name': 'Participant Name',
                          'Leader Email':'Participant Email',
                          'Leader Phone':'Participant Phone Number' ,"Candidate Type":'Participant Type'})
df1.head()
```

```python
def member_extractor(df):
    df['Members'] = df['Members'].str.split(', ')
    df3 = df[df['Members'].apply(len) == 1].copy()
    df4 = df[df['Members'].apply(len) > 1].copy()
    
    return df3, df4

df3, df4 =member_extractor(df1)
```

```python
df3['Members'] = df3['Members'].apply(lambda x: ' '.join(x))

df3.drop(['Members'], axis=1,inplace=True)
```

```python
df4.head()
```

```python
import pandas as pd

def func(df):
    new_rows = []
    for index, row in df.iterrows():
        team_name = row['Team Name']
        leader_email = row['Participant Email']
        members = row['Members']
        for member_email in members[1:]:
            new_rows.append({
                'Team Name': team_name,
                'Participant Name': '',
                'Participant Email': member_email,
                'Participant Phone Number': '',
                'Members': [],
                'Participant Type': 'team member'
            })
    new_df = pd.concat([df] + [pd.DataFrame(new_rows)], ignore_index=True)
    return new_df
df4 = func(df4)
```

```python
df4.drop(['Members'], axis=1,inplace=True)
```

```python
df= pd.concat([df2, df3,df4])
```
