```python
df2=pd.read_csv('/kaggle/input/ieee-jusb/anon_data2.xlsx - Sheet1 (1).csv')
```

```python
df2.drop(columns=['Registration Time','Passing Out Year','Course Pursuing',"Candidate's Gender",'Course Stream','Country','Specialization'],inplace=True)
```

```python
df2.head()
```

```python
member = df2[df2['Candidate Type'] != 'team leader']
member.head(20)
```

```python
def combine(member):
    member_email = member.groupby('Team Name')["Candidate's Email"].apply(','.join).reset_index()
    member_email.columns = ['Team Name', 'members']
    return member_email
member_df = combine(member)
member_df
```

```python
df3 = df2.copy()
df3.drop(columns='Team Id', inplace=True)
df3 = df3[df3['Candidate Type'] != 'team member']
```

```python
# Merge df3 with member_df on 'Team Name' with default suffixes
df3 = df3.merge(member_df, on='Team Name', how='left')

df3.head()
```

```python
def combine2(df3):
    df3.loc[df3['Candidate Type'] == 'team leader', 'members'] = df3["Candidate's Email"]+','+df3['members']
    return df3

# Apply the function to df3
df3 = combine2(df3)
df3.head()
```

```python
def f(df3):
    x = df3['members'].isnull()
    df3.loc[x, 'members'] = df3.loc[x, "Candidate's Email"]
    return df3
f(df3)
df3.head()
```

```python
def replace_year(df):
    year_mapping = {
        'First Year': 1,
        'Second Year': 2,
        'Third Year': 3,
        'Fourth Year': 4
    }
    for year, value in year_mapping.items():
        df.loc[df['Year Of Study'] == year, 'Year Of Study'] = value
replace_year(df3)
df3.drop(columns=['Candidate Type'], inplace=True)
df3.head(5)
```

```python
df1.head(5)
```

```python
df3 = df3.rename(columns={"Candidate's Name": 'Leader Name',
                          "Candidate's Email":'Leader Email',
                          "Candidate's Mobile":'Leader Phone' ,
                          'Year Of Study': 'Leader Year',
                          "Candidate's Organisation": 'Leader College',
                          'members':'Members'})

df3.head()
```

```python
df3.shape[0]+df1.shape[0]
```

```python
concatenated_df = pd.concat([df3, df1])
concatenated_df.head()
```

```python
concatenated_df.shape
```

```python
a = concatenated_df['Team Name'].duplicated().any()
b= concatenated_df['Leader Name'].duplicated().any()

# Print the results
print("Are there any duplicate Team Names?", a)
print("Are there any duplicate Leader Names?", b)
```

```python
aa= concatenated_df[concatenated_df['Team Name'].duplicated()]['Team Name']
bb= concatenated_df[concatenated_df['Leader Name'].duplicated()]['Leader Name']

print("Duplicate Team Names:")
print(aa)

print("Duplicate Leader Names:")
print(bb)
```

```python
concatenated_df.drop_duplicates(subset=['Team Name'], keep='first', inplace=True)
concatenated_df.shape
```
