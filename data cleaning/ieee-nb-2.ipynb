```python
df1=pd.read_csv('/kaggle/input/ieee-jusb/anon_data1.xlsx - Sheet1 (1).csv')
df2=pd.read_csv('/kaggle/input/ieee-jusb/anon_data2.xlsx - Sheet1 (1).csv')
```

```python
df2.drop(columns=['Registration Time','Passing Out Year','Course Pursuing',"Candidate's Gender",'Course Stream','Country','Specialization','Year Of Study',"Candidate's Organisation"],inplace=True)
```

```python
df2.drop(['Team Id'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
df1.drop(['Leader College','Leader Year','Members'], axis =1 , inplace=True)
df1['Candidate Type']='team leader'
```

```python
df1.head()
```

```python
df2 = df2.rename(columns={"Candidate's Name": 'Participant Name',
                          "Candidate's Email":'Participant Email',
                          "Candidate's Mobile":'Participant Phone Number' ,"Candidate Type":'Participant Type'})
```

```python
df1 = df1.rename(columns={'Leader Name': 'Participant Name',
                          'Leader Email':'Participant Email',
                          'Leader Phone':'Participant Phone Number' ,"Candidate Type":'Participant Type'})
```

```python
df2.head()
```

```python
df1.head()
```

```python
df3 = pd.concat([df2, df1])
df3.head()
```

```python
df3.shape
```

```python
a = df3['Team Name'].duplicated().any()
b= df3['Participant Name'].duplicated().any()

# Print the results
print("Are there any duplicate Team Names?", a)
print("Are there any duplicate Participant Names?", b)
```

```python
aa= df3[df3['Team Name'].duplicated()]['Team Name']
bb= df3[df3['Participant Name'].duplicated()]['Participant Name']

print("Duplicate Team Names:")
print(aa)

print("Duplicate  Participant Names:")
print(bb)
```

```python
df3.drop_duplicates(subset=['Participant Name'], keep='first', inplace=True)
df3.shape
```

```python
df3.to_csv('/kaggle/working/output.csv', index=False)
```
